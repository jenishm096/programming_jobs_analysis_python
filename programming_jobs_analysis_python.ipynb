```python
# Checking value counts for workloc column
workloc_value_counts=df['WorkLoc'].value_counts()
print(workloc_value_counts)
workloc_value_counts.index=['Office','Home','Other']
print('-'*50)
print(workloc_value_counts)
#plotting the data
plt.figure(figsize=(15,6))
workloc_value_counts.plot(kind='barh',color='orange')
plt.xlabel('Work Location')
plt.ylabel('Number of Respondent')
plt.title('Number of Respondent by Work Location')
plt.gca().invert_yaxis()
plt.show()
```

```python
#Replacing null values with a majority values
df['WorkLoc'].fillna('Office',inplace=True)
```

```python
#Listing out various categories in 'CompFreq'
df['CompFreq'].unique()
```

```python
#Creating 'AnnualComp' column using 'TotalComp' and 'CompFreq' columns
df['AnnualComp']=0
for i,r in df.iterrows():
    if r['CompFreq']=='Monthly':
        df.loc[i,'AnnualComp']=r['CompTotal']*12
    elif r['CompFreq']=='Weekly':
        df.loc[i,'AnnualComp']=r['CompTotal']*52
    else:
        df.loc[i,'AnnualComp']=r['CompTotal']
```

```python
#Finding the value of quartile range of column 'ConvertedComp'
Q1=df['ConvertedComp'].quantile(0.25)
median=df['ConvertedComp'].median()
Q3=df['ConvertedComp'].quantile(0.75)
print(f'The Q1, median and Q3 of the column ConvertedComp are {Q1}, {median} and {Q3}.')
```

```python
#Calculating Inter quartile range of ConvertedComp
IQR=Q3-Q1
print('The interquartile range of converted comp is:',IQR)
```

```python
#Finding minimum and maximum values
min_val = df['ConvertedComp'].min()
max_val = df['ConvertedComp'].max()
print('The min/max for ConvertedComp are',min_val, 'and', max_val)
```

```python
#Finding outliers in the column 'ConvertedComp'
upper_outliers = df[df['ConvertedComp']>(Q3+IQR*1.5)]['ConvertedComp']
lower_outliers = df[df['ConvertedComp']<(Q1-IQR*1.5)]['ConvertedComp']


#calculating number of outliers
print('Outlier Below:',lower_outliers.count())
print('Outlier Above:',upper_outliers.count())
print(f'There are {upper_outliers.count()+lower_outliers.count()} number of outliers in column ConvertedComp.')
print('-'*50)
print('Median with outliers:',df['ConvertedComp'].median())
print('Median without outliers:',df[df['ConvertedComp']<(Q3+IQR*1.5)]['ConvertedComp'].median())
print('Mean with outliers:',df['ConvertedComp'].mean().round(2))
print('Mean without outliers:',df[df['ConvertedComp']<(Q3+IQR*1.5)]['ConvertedComp'].mean().round(2))
```

```python
#removing rows with outliers in 'ConvertedComp' Columns
df1=df[df['ConvertedComp']<(Q3+IQR*1.5)]
print('New dataset shape:',df1.shape)
```
